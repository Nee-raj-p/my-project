```python
## To remove duplicate values 
hb.drop_duplicates(inplace=True)
uni_rows=hb.shape[0]
uni_rows
```

```python
per_comp = rem_val[0]/uni_rows*100
print(per_comp)
hb.drop(['company'],axis=1,inplace=True)
# better to drop company because more missing values
```

```python
per_ag = rem_val[1]/uni_rows*100
print(per_ag)
hb['agent'].fillna(value=0,inplace=True)
# better to change to 0 because relatively less number of missing values
```

```python
per_cou = rem_val[2]/uni_rows*100
print(per_cou)
hb['country'].fillna(value='others',inplace=True)
```

```python
per_ch = rem_val[3]/uni_rows*100
print(per_ch)
hb['children'].fillna(value=0,inplace=True)
```

```python
hb.isnull().sum()
```

```python
hb[['children','agent']] = hb[['children','agent']].astype('int64')
```

* Average Daily Rate, as defined by dividing the sum of all lodging transactions by the total number of staying nights

## Data Filtering and Grouping 

```python
hb['total_stay_in_night'] = hb['stays_in_week_nights'] + hb['stays_in_weekend_nights']
hb['total_stay_in_night']
# created new column and grouping to reduce more columns
```

```python
# created a col for revenue using total stay * adr
hb['revenue'] = hb['total_stay_in_night'] * hb['adr']
hb['revenue']
```

```python
# add a column with total guest coming for each booking
hb['total_guest'] = hb['adults'] + hb['children'] + hb['babies']
hb['total_guest'].sum()
```

```python
# we will replace the value from (0,1) to not_canceled, is canceled.

hb['is_canceled'] = hb['is_canceled'].replace([0,1], ['not canceled', 'is canceled'])
hb['is_canceled']
```

```python
# total revenue 
hotel_rev = hb.groupby('hotel')['revenue'].sum()
hotel_rev
```

## Data Visualisation

```python
hb['hotel'].value_counts().plot.pie(explode=[0.05,0.05],autopct='%1.1f%%',shadow=True,figsize=(10,8),fontsize=25)
plt.title('Most Preferred Hotel',fontsize=20)
```

```python
grp_htl = hb.groupby('hotel')
high_adr = grp_htl['adr'].mean().reset_index()
plt.figure(figsize=(7,5))
plt.xlabel('Hotel Type',fontsize=18)
plt.ylabel('ADR',fontsize=18)
plt.title('Average ADR of Hotel Type',fontsize=21)
sns.barplot(x=high_adr['hotel'],y=high_adr['adr'],width=0.4)
```

```python
hb['is_repeated_guest'].value_counts().plot.pie(explode=[0.05,0.05],autopct='%1.1f%%',figsize=(7,5),fontsize=12)
plt.title('Repeated Guest Percentage',fontsize=18)
```

```python
plt.figure(figsize=(9,7))
sns.countplot(x=hb['meal'])
plt.xlabel('Meal Type',fontsize=12)
plt.ylabel('Count',fontsize=12)
plt.title('Most Preferable Meal Type',fontsize=18)
```

```python
dist_ch = hb.groupby(['distribution_channel','hotel'])['adr'].mean().reset_index()
plt.figure(figsize=(17,9))
sns.barplot(x='distribution_channel',y='adr',data=dist_ch,hue='hotel')
plt.title('ADR Distribution Across Channels',fontsize=21)
```

```python
canc_perc = hb['is_canceled'].value_counts(normalize=True)
print(canc_perc)
plt.figure(figsize = (9,7))
plt.title('Reservation status count')
plt.bar(['Not canceled','Canceled'],hb['is_canceled'].value_counts(),edgecolor='k',width=0.4)
plt.show()
```

```python
plt.figure(figsize=(5,5))
ax1= sns.countplot(x='hotel',hue='is_canceled',data=hb,palette='Blues')
plt.title('Reservation status in different hotels',size=14)
plt.xlabel('Hotel')
plt.ylabel("Number Of Reservations")
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.countplot(x=hb['assigned_room_type'],order=hb['assigned_room_type'].value_counts().index)
plt.xlabel('Room Type',fontsize=18)
plt.ylabel('Count of Room Type',fontsize=18)
plt.title('Most Preferred Room Type',fontsize=23)
```

```python
plt.figure(figsize = (8,5))
hotel_wise_revenue = hb.groupby('hotel')['revenue'].sum()
hotel_wise_revenue
ax = hotel_wise_revenue.plot(kind = 'bar', color = ('c', 'm'),width=0.4)
plt.xlabel("Hotel", fontdict={'fontsize': 12, 'fontweight' : 5, 'color' : 'Brown'})
plt.ylabel("Total Revenue", fontdict={'fontsize': 12, 'fontweight' : 5})
plt.title("Total Revenue Made By Hotels", fontdict={'fontsize': 12, 'fontweight' : 5})
```

```python
plt.figure(figsize = (12,6))
sns.scatterplot(y = 'total_stay_in_night', x = 'adr', data = hb[hb['adr'] < 1000])
plt.show()
# guest stay days decreasing , ADR gets high
```

```python
corr_df = hb[['lead_time','previous_cancellations', 'previous_bookings_not_canceled', 'total_guest',
                    'booking_changes', 'days_in_waiting_list', 'adr', 'required_car_parking_spaces', 'total_of_special_requests']].corr()
plt.figure(figsize=(12,12))
sns.heatmap(corr_df, annot = True, fmt='.2f', annot_kws={'size': 12}, square=True, cmap="YlGnBu")
```
